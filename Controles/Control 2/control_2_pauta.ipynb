El juego es de carácter casual y será lanzado en Android, iOS y en la consola _Nientiendo Swap_.

Una partida de Bolsimon Unión consiste en dos equipos de 5 jugadores (azul y rojo) que se enfrentan durante 10 minutos con el objetivo de conseguir la mayor cantidad de puntos posible. Las partidas se caracterizan por lo siguiente:

* Cada jugador elige un bolsimon de los que tenga disponibles en su _pool_ de bolsimones. Un bolsimon no puede estar dos veces en una misma partida.
* A lo largo de la partida los jugadores deben dejar fuera de combate (KO) a los bolsimones oponentes para avanzar hacia su base y así poder obtener puntos (_score_).
* Cuando más de un jugador participa en dejar KO a un oponente, aquellos que no dieron el último golpe reciben una asistencia.
* Un bolsimon KO vuelve a la partida luego de unos segundos.
* El equipo ganador es aquel que logró obtener más puntos luego de 10 minutos de partida.
* Para las estadísticas el juego guarda el detalle de cada KO y cada _score_.

1. Para el usuario de email `elizabeth@cooper.net` encontrar la cantidad total de KOs, Asistencias y puntaje obtenidos en todas sus partidas.

```python
%%sql
WITH elizabeth_id AS (
  SELECT usuario.id AS id
  FROM usuario
  WHERE usuario.email = 'elizabeth@cooper.net'
)
SELECT ko.ko_count, asistencia.assist_count ,score.score_count
FROM
  (SELECT COUNT(*) AS ko_count
   FROM ko, elizabeth_id
   WHERE ko.id_jugador = elizabeth_id.id
  ) AS ko,
  (SELECT SUM(score.puntaje) AS score_count
   FROM score, elizabeth_id
   WHERE score.id_jugador = elizabeth_id.id
  ) AS score,
  (SELECT COUNT(*) AS assist_count
   FROM asistencia, elizabeth_id
   WHERE asistencia.id_jugador = elizabeth_id.id
  ) AS asistencia
```

2. Una consulta que entregue el equipo ganador de la partida de id `12345`

```python
%%sql
SELECT upb.equipo
FROM partida
JOIN usuario_partida_bolsimon AS upb ON upb.id_partida = partida.id
JOIN score on score.id_partida = partida.id AND score.id_jugador = upb.id_jugador
WHERE partida.id = 12345
GROUP BY upb.equipo
ORDER BY SUM(score.puntaje) DESC
LIMIT 1

```

5. Los ids de la pareja de jugadores que más asistencias se hayan hecho mutuamente. (Es decir si `A` asistió a `B` 5 veces y `B` a `A` 3 veces el total es 8)

```python
%%sql
WITH pair_assist_counts AS(
  SELECT ko.id_jugador AS koer, asistencia.id_jugador AS assister, COUNT(*)
  FROM ko JOIN asistencia on ko.id = asistencia.id_ko
  GROUP BY koer, assister
)
SELECT p1.koer, p1.assister, (p1.count + p2.count) AS total_count
FROM pair_assist_counts AS p1, pair_assist_counts AS p2
WHERE p1.koer = p2.assister AND p1.assister = p2.koer
ORDER BY total_count DESC
LIMIT 1
```

6. Encontrar el username del usuario que jugó más partidas en un día. Tip: Con la sentencia `date_trunc('day', <fecha>)` se puede extraer el día a partir de una fecha completa.

```python
%%sql
SELECT usuario.username, usuario.id 
FROM
  (SELECT usuario.id AS id, usuario.username AS username, COUNT(partida.id) AS cuenta
  FROM usuario
  JOIN usuario_partida_bolsimon AS upb ON upb.id_jugador = usuario.id
  JOIN partida ON upb.id_partida = partida.id
  GROUP BY usuario.id, date_trunc('day', partida.fecha_inicio)) AS agg
JOIN usuario ON usuario.id=agg.id
GROUP BY usuario.id
ORDER BY MAX(agg.cuenta) DESC
LIMIT 1
```

7. Encontrar el Bolsimon, equipo, la cantidad KOs realizados y puntos totales de cada jugador de la partida de ID `32`. Los jugadores del equipo ganador deben aparecer primero y los jugadores de cada equipo deben ir ordenados por puntos totales y KOs realizados de manera descendente.

```python
%%sql

with winning_team AS (
  SELECT upb.equipo
  FROM usuario_partida_bolsimon AS upb
  JOIN score on score.id_partida = upb.id_partida AND score.id_jugador = upb.id_jugador
  WHERE upb.id_partida = 32
  GROUP BY upb.equipo
  ORDER BY SUM(score.puntaje) DESC
  LIMIT 1
), kos AS (
  SELECT ko.id_jugador AS id_jugador, COUNT(*) AS ko_count
  FROM ko JOIN usuario_partida_bolsimon AS upb ON upb.id_jugador = ko.id_jugador
  AND upb.id_partida = ko.id_partida
  WHERE upb.id_partida=32
  GROUP BY ko.id_jugador
), scores AS (
  SELECT score.id_jugador AS id_jugador, SUM(score.puntaje) AS score_count
  FROM score JOIN usuario_partida_bolsimon AS upb ON upb.id_jugador = score.id_jugador
  AND upb.id_partida = score.id_partida
  WHERE upb.id_partida=32
  GROUP BY score.id_jugador
)

SELECT upb.equipo, bolsimon.nombre, COALESCE(kos.ko_count, 0) AS kos, COALESCE(scores.score_count, 0) AS total_score, upb.id_jugador
FROM winning_team, usuario_partida_bolsimon AS upb
JOIN bolsimon ON bolsimon.id = upb.id_bolsimon
LEFT JOIN kos on kos.id_jugador = upb.id_jugador
LEFT JOIN scores on scores.id_jugador = upb.id_jugador
WHERE upb.id_partida = 32
ORDER BY upb.equipo = winning_team.equipo DESC, total_score DESC, kos DESC
```

8. Encontrar el nombre del Bolsimon de mejor y peor winrate (% de victorias).Muestre los winrates de cada uno (basta con mostrar 2 filas en que una corresponde al mejor y otra al peor).

**Protips**:
- Para contar la cantidad de `true` en una columna (o agregación) hay que castear los booleanos a entero (`1` o `0`) y despues sumar, algo así:
`SUM(CAST(<expresion o columna booleana> AS INTEGER)`.
- Para hacer división con decimales en postgres es necesario que al menos uno de los dos términos sea de tipo `FLOAT`, es decir podemos hacer `SELECT CAST(1 AS FLOAT) / 2` y eso da `0.5`.

```python
%%sql
with team_scores AS (
  SELECT upb.id_partida, upb.equipo, SUM(score.puntaje) AS puntaje
  FROM usuario_partida_bolsimon AS upb
  JOIN score on score.id_partida = upb.id_partida AND score.id_jugador = upb.id_jugador
  GROUP BY upb.id_partida, upb.equipo
), 
winning_teams AS (
  SELECT team_scores.id_partida, team_scores.equipo
  FROM team_scores JOIN (
    SELECT t2.id_partida, MAX(t2.puntaje) AS puntaje
    FROM team_scores AS t2
    GROUP BY t2.id_partida
  ) AS t_max ON t_max.id_partida = team_scores.id_partida
  AND t_max.puntaje = team_scores.puntaje
)
(SELECT bolsimon.nombre, SUM(CAST(upb.equipo = winning_teams.equipo AS INTEGER)::FLOAT) / COUNT(*) * 100 AS win_rate
FROM usuario_partida_bolsimon AS upb
JOIN bolsimon ON bolsimon.id = upb.id_bolsimon
JOIN winning_teams ON winning_teams.id_partida = upb.id_partida
GROUP BY bolsimon.id, bolsimon.nombre
ORDER BY win_rate DESC
LIMIT 1)
UNION
(SELECT bolsimon.nombre, SUM(CAST(upb.equipo = winning_teams.equipo AS INTEGER)::FLOAT) / COUNT(*) * 100 AS win_rate
FROM usuario_partida_bolsimon AS upb
JOIN bolsimon ON bolsimon.id = upb.id_bolsimon
JOIN winning_teams ON winning_teams.id_partida = upb.id_partida
GROUP BY bolsimon.id,  bolsimon.nombre
ORDER BY win_rate ASC
LIMIT 1)
```
